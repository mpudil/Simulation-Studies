## Exercise in Permutation (Pasting)

Goal: Perform a one-sided permutation test (alternative hypothesis is that the ratio is greater than 1) on the ratio of variances (i.e., var(ILEC)/var(CLEC)). 

```r

ILEC <-c(1,1,1,1,2,2,1,1,1,1,2,2,1,1,1,1,2,2,1,1,1,1,2,3,1,1,1,1,2,3,1,1,1,1,2,3,1,1,1,1,2,3,1,1,1,1,2,3,1,1,1,1,2,3,1,1,1,1,2,4,1,1,1,1,2,5,1,1,1,1,2,5,1,1,1,1,2,6,1,1,1,1,2,8,1,1,1,1,2,15,1,1,1,2,2) 
CLEC <- c(1,1,5,5,5,1,5,5,5,5) 

stat <- (var(ILEC) / var(CLEC))
yobs <- c(ILEC, CLEC)
N <- length(yobs)
n1 <- length(ILEC)
n2 <- length(CLEC)
nperms <- 10000
Fstat <- NULL

for(i in 1:nperms) {
  grp1 <- sample(1:N, n1, replace=FALSE)
  Fstat[i] <- (var(yobs[grp1]) / var(yobs[-grp1]))
  
}

hist(Fstat, col='blue')
abline(v=stat, lwd=3, col='red')
```

### Take observed statistic, and see what percentage of them are more or as extreme as our statistic



Including the negative of this doesn't really make sense since we're talking about variances, which are
always positive

### Calculate p value


```r
p.value <- sum(stat < Fstat)/nperms 
print(noquote(paste('The p-value is', p.value)))
```

Again, including the negative extreme will not make a difference since all values are positive. 

 The p value is is the probability of finding the observed, or more extreme, results when H0 is true. Here, since the pvalue is about 0.26, we find that there is a 81.7% chance of finding our ratio between the variances of  ILEC and CLEC, or a higher variance, if the null hypothesis is true (that is, if there really is no difference between the ratio of
ILEC and CLEC variances). 


The first P value associated with the F test (0.66) is under the assumption that the data comes from a normal distribution. The p value associated with the permutation test, however, is not under the assumption of normality, but only under the assumption that the two groups (ILEC and CLEC)come from the same distribution, whatever that may be. Now, it seems that the ends of the distributions is where the majority of the data is, so it is likely that the permutation test will be more accurate than the F test.
